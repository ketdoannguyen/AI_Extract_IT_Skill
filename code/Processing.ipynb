```python
import pandas as pd
import re
```

```python
df_read = pd.read_csv("./data/JobDescription.csv",sep="\t")
df_read.info()
```

```python
list_JDs = list(df_read.JD)
list_JDs = list(dict.fromkeys(list_JDs))
# list_JDs = [job for job in list_JDs if len(job)>300 and len(job)<=6000]
len(list_JDs)
```

```python
for i in range(len(list_JDs)):
    while True :
        check = re.search("[^ ]/ ",list_JDs[i])
        if check:
            list_JDs[i] = list_JDs[i][:check.start()+2] + list_JDs[i][check.start()+3:]
            print(list_JDs[i][check.start()-3:check.start()+3])
        else:
            break
```

```python
for i in range(len(list_JDs)):
    while True :
        check = re.search("[^a-zA-Z0-9 :/]/",list_JDs[i])
        if check:
            list_JDs[i] = list_JDs[i][:check.start()+1] +" / "+ list_JDs[i][check.start()+2:]
            print(list_JDs[i][check.start()-5:check.start()+10])
        else:
            break
    while True :
        check = re.search(" /[^ ]",list_JDs[i])
        if check:
            list_JDs[i] = list_JDs[i][:check.start()+1] +"/ "+ list_JDs[i][check.start()+2:]
            print(list_JDs[i][check.start()-5:check.start()+10])
        else:
            break
```

```python
df_save = pd.DataFrame({"JD":list_JDs})
df_save.to_csv("./data/JobDescription.csv",index=False)
```
