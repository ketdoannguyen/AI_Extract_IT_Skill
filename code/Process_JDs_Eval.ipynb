```python
import spacy
import re
import numpy as np
import pandas as pd
from IT_Skill_NER_Class import IT_SKILL_NER_Class
```

```python
df = pd.read_csv("/home/vkuai/AI_Extract_IT_Skill/data/JDs_Eval.csv", sep=";")
list_JDs = list(df["nameJob"]+". "+df["JDs"]+".")
print(len(list_JDs))
list_JDs[:5]
```

```python
model = IT_SKILL_NER_Class()
list_skills = []
for jd in list_JDs:
    doc = model.nlp(jd)
    skill = model._get_hard_skill(doc)
    list_skills.append(skill)
```

```python
average = np.sum([len(skill) for skill in list_skills])/len(list_skills)
print(f"JDs Evaluating has {str(len(list_skills))} JDs")
print(f"Each JD has an average of {str(average)} skill")
```

```python

df['skill'] = list_skills
df.to_csv("./data/JDs_Eval.csv",sep=";",index=False)
```
